```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= pd.read_csv("final_project_data.csv") # call data
```

```python
data['DATE'] = pd.to_datetime(data['DATE']) # change DATE column to datetime objects
# Last point is from Jan 1st, 2026. Unable to make conclusions about 2026 with just one point, so will focus up until end of 2025.
# Also extract only the two necessary columns
df1 = data.loc[:len(data)-2, ['DATE','SNOW']]
```

```python
by_month=df1.groupby(pd.Grouper(key='DATE', axis=0, freq='ME')).sum() # Find amount of snow/month to condense data.

by_month = by_month.reset_index() # change so dataframe indices aren't dates
by_month
```

```python
# scatterplot
alt.Chart(by_month).mark_circle().encode(x="DATE", y="SNOW", size='SNOW').properties(title="Total Monthly Snowfall")
```

```python
by_month['YEAR'] = by_month['DATE'].dt.year 
selection = alt.selection_point(fields=['YEAR'], bind='legend')

alt.Chart(by_month).mark_area(opacity=1,line=True).encode(
    x = alt.X('month(DATE):T', title='Month'),
    y = alt.Y('SNOW:Q', title='Total Snowfall (in)', stack=False),
    color=alt.Color('YEAR:N', title='Year', scale=alt.Scale(scheme='lightgreyteal')),
    tooltip=['YEAR'],
    opacity=alt.when(selection).then(alt.value(1)).otherwise(alt.value(0.05))
).add_params(selection).properties(title='Snowfall Throughout Each Year',width=800).interactive()
```

```python
alt.Chart(by_month).mark_line().encode(
    x = alt.X('year(DATE):T', title=None),
    y = alt.Y('SNOW:Q', title='Total Snowfall (in)'),
    opacity=alt.condition(selection,alt.value(1),alt.value(.2)),
).add_params(
    selection).interactive(
).properties(
    width=90, height=200
).facet(
    column='month(DATE):N', title = "Snowfall Trends through the Years Within Each Month")
```
